### 11a. Venues for high scoring games

```python
match['first_innings'] = match['target_runs'] - 1

match['second_innings_runs'] = np.where(
    match['result'] == 'runs',
    match['first_innings'] - match['result_margin'],
    match['target_runs']  
)

# Total match runs
match['total_runs'] = match['first_innings'] + match['second_innings_runs']

high_scoring = match[match['total_runs'] >= 400]

venue_high_scores = high_scoring['venue'].value_counts()

venue_high_scores.head(3)
```

```python
super_close=match[((match['result_margin']<10) & (match['result']=='runs')) | ((match['result_margin']<3) & (match['result']=='wickets'))]
super_close['venue'].value_counts().head(3)
```
