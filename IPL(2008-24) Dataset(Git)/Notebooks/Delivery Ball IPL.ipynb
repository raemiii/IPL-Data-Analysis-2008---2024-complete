### 2a. Strike Rate Calculation

```python
total_runs_batter = delivery.groupby('batter')['batsman_runs'].sum().reset_index(name='runs')
total_runs_batter
```

```python
balls_faced=delivery[~delivery['extras_type'].isin(['wides','noballs'])].groupby('batter').size().reset_index(name='balls')
balls_faced
```

```python
stats=pd.merge(total_runs_batter,balls_faced,on='batter',how='inner')
stats
```

```python
stats['SR']=(stats['runs']/stats['balls'])*100
stats['SR']
```

```python
high_SR=stats[stats['balls']>=100].sort_values(by='SR',ascending=False).head(10)
high_SR
```

### 2b.Top Scorers

```python
top_scorers=delivery.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)
top_scorers
```

### 2c. Highest Individual Scores

```python
individual_score=delivery.groupby(['batter','match_id'])['batsman_runs'].sum().sort_values(ascending=False).head(10)
individual_score
```

### 2d. Fastest 50s

```python
final_runs=delivery.groupby(['match_id', 'batter'])['batsman_runs'].sum().reset_index(name='runs')
final_runs
```

```python
balls_faced=delivery[~delivery['extras_type'].isin(['wides', 'noballs'])].groupby(['match_id', 'batter']).size().reset_index(name='balls')
balls_faced
```

```python
ind_inn_score=pd.merge(final_runs, balls_faced, on=['match_id', 'batter'], how='inner')
ind_inn_score
```

```python
ind_inn_score['sr']=(ind_inn_score['runs'] / ind_inn_score['balls']) * 100
ind_inn_score['sr']
```

```python
fastest_50=ind_inn_score[ind_inn_score['runs'] >= 50].sort_values(by='sr', ascending=False).head(10)
fastest_50
```

```python
fastest_100=ind_inn_score[ind_inn_score['runs'] >= 100].sort_values(by='sr', ascending=False).head(10)
fastest_100
```

### 3e. Best Economy (across seasons)

```python
total_dots_bowler = delivery[(delivery['total_runs'] == 0)].groupby('bowler').size().sort_values(ascending=False).reset_index(name='runs')
total_dots_bowler
```

```python
total_runs_bowler = delivery.groupby('bowler')['total_runs'].sum().reset_index(name='runs')
total_runs_bowler
```

```python
balls_bowled=delivery[~delivery['extras_type'].isin(['wides','noballs'])].groupby('bowler').size().reset_index(name='balls')
balls_bowled
```

```python
eco_stats=pd.merge(total_runs_bowler,balls_bowled,on='bowler',how='inner')
eco_stats
```

```python
eco_stats=pd.merge(eco_stats,total_wkts_bowler,on='bowler',how='inner')
eco_stats
```
